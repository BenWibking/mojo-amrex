```python
import asyncio
import io
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
from ipywidgets import Image, Label, VBox
from sidecar import Sidecar

import mojo.importer
import heat_equation_vis
```

```python
def png_bytes(array):
    buffer = io.BytesIO()
    plt.imsave(buffer, array, cmap='viridis', vmin=1.0, vmax=2.0, origin='lower', format='png')
    return buffer.getvalue()

sc = Sidecar(title='Heat Equation')
step_widget = Label(value='Step: 0')
image_widget = Image(format='png')
image_widget.value = png_bytes(np.ones((128, 128)))

with sc:
    display(VBox([step_widget, image_widget]))
```

```python
runner = heat_equation_vis.HeatEquationRunner()
image_widget.value = png_bytes(np.array(runner.slice_array(), copy=True))

step_count = 0
step_widget.value = f'Step: {step_count}'
while runner.step():
    step_count += 1
    step_widget.value = f'Step: {step_count}'
    image_widget.value = png_bytes(np.array(runner.slice_array(), copy=True))
    await asyncio.sleep(0.01)

print(f'Completed {step_count} steps')
```
